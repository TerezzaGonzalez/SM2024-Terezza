**3. Crea una función que cree una gráfica y dibuje 'k' líneas de f(x)=Sin( $\alpha$ x). La función recibe:**

* x1: valor inicial de 'x'
* x2: valor final de 'x'
* n: cantidad de puntos del vector 'x'
* a: valor inicial de $\alpha$
* b: valor final de $\alpha$
* k: cantidad de líneas

Usar el la variable opcional de "label" y el comando de legend para saber qué línea pertenece a qué $\alpha$.

```python
import numpy as np
import matplotlib.pyplot as plt

def función(x1, x2, n, a, b, k):
    x_valores = np.linspace(x1, x2, n) #Valores de x
    alfa_valores = np.linspace(a, b, k) # valores de alfa
    
    for alfa in alfa_valores: #Iteración 
        y = np.sin(alfa * x_valores) #Función que nos pidio
        plt.plot(x_valores, y, label=f'α = {alfa:.2f}') # Esto lo investigue porque no sabía hacer la etiqueta de los 4 alfas
        
    plt.grid()    
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
   
función(-2*np.pi, 2*np.pi, 100, 1, 4, 4)
```
